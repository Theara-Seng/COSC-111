#### 1. Write a Python program to:
#### • Find and print the name(s) of the student(s) with the highest score.
#### • Calculate and print the average score of the class.
#### • Create a new dictionary that only includes students who scored at least 80.


```python
scores = {
    'Alice': 85,
    'Bob': 92,
    'Charlie': 78,
    'David': 90,
    'Eve': 88
}
# Find and print the name(s) of the student(s) with the highest score.
max_score = max(scores.values())
for name, score in scores.items():
    if score == max_score:
        print("Top student:", name)
# Calculate and print the average score of the class.
total = 0
for score in scores.values():
    total += score
average = total / len(scores)
print("Average score:", average)

# Create a new dictionary that only includes students who scored at least 80.
scores_atleast_80 = {}
for name, score in scores.items():
    if score >= 80:
        scores_atleast_80[name] = score

print("Students with score >= 80:", scores_atleast_80)

```

#### 2. Write a program to:
#### • Find the top 3 students with the highest scores and print their names and scores.

```python
scores = {
'Alice': 85, 'Bob': 92, 'Charlie': 78,
'David': 90, 'Eve': 88, 'Frank': 95, 'Grace': 80
}
sorted_scores = sorted(scores.items(), key=lambda x: x[1], reverse=True)
top_3 = sorted_scores[:3]

print("Top 3 students:")
for name, score in top_3:
    print(f"{name}: {score}")
```

#### 3. Write a program to:
#### • Compute and print the total sales and total expenses for the quarter.
#### • Also calculate the profit (sales - expenses) for each month.

```python
data = {
    'January': {'sales': 1000, 'expenses': 400},
    'February': {'sales': 1200, 'expenses': 500},
    'March': {'sales': 900, 'expenses': 450}
}

#  Compute and print the total sales and total expenses for the quarter.
total_sales = 0
total_expenses = 0

for month, values in data.items():
    total_sales += values['sales']
    total_expenses += values['expenses']

print("Total Sales for the Quarter:", total_sales)
print("Total Expenses for the Quarter:", total_expenses)

#  Also calculate the profit (sales - expenses) for each month.
for month, values in data.items():
    profit = values['sales'] - values['expenses']
    print(month, "Profit:", profit)
```

#### 4. Write a program to create a dictionary where the keys are name lengths and the values are lists of names of that length:

```python
names = ["Anna", "Benjamin", "Cathy", "Dann", "Eleanor", "Frank"]

name_lengths = {}

for name in names:
    length = len(name)
    
    if length in name_lengths:
        name_lengths[length].append(name)
    else:
        name_lengths[length] = [name]

print("Names grouped by length:")
for key, value in name_lengths.items():
    print(key, ":", value)
```

#### 5. Build a simple console-based student management system that uses Python dictionaries to store and manage data about students, their courses, and grades.

```python
students = {
    'S001': {
        'name': 'Alice',
        'courses': {'Math': 90, 'English': 85}
    },
    'S002': {
        'name': 'Bob',
        'courses': {'Math': 75, 'Science': 80}
    }
}

def add_student(student_id, name, courses_dict):
    if student_id in students:
        print(f"Student ID {student_id} already exists.")
    else:
        students[student_id] = {
            'name': name,
            'courses': courses_dict
        }
        print(f"Student '{name}' with ID {student_id} added successfully.")

def add_update_grade(student_id, course, grade):
    if student_id in students:
        students[student_id]['courses'][course] = grade
        print(f"Added/Updated {course} for {students[student_id]['name']}: {grade}")
    else:
        print(f"Student ID {student_id} not found!")
            
def remove_student(student_id):
    if student_id in students:
        name = students[student_id]['name']
        del students[student_id]
        print(f"Removed student: {name} ({student_id})")
    else:
        print(f"Student ID {student_id} not found!")

def display_all_students():
    print("\nAll students = {")
    for sid, info in students.items():
        print(f"  '{sid}': {{")
        print(f"    'name': '{info['name']}',")
        print(f"    'courses': {info['courses']}")
        print("  },")
    print("}")

def show_average(student_id):
    if student_id in students:
        student = students[student_id]
        if student['courses']:
            grades = student['courses'].values()
            average = sum(grades) / len(grades)
            print(f"{student['name']}'s average: {average:.2f}")
        else:
            print(f"{student['name']} has no courses yet!")
    else:
        print(f"Student ID {student_id} not found!")

def find_top_student():
    top_score = 0
    top_students = []
    for sid, info in students.items():
        grades = info['courses'].values()
        if grades:
            avg = sum(grades) / len(grades)
            if avg > top_score:
                top_score = avg
                top_students = [(sid, info['name'], avg)]
            elif avg == top_score:
                top_students.append((sid, info['name'], avg))
    if top_students:
        for sid, name, avg in top_students:
            print(f"ID: {sid}, Name: {name}, Average: {round(avg, 2)}")
    else:
        print("No students have grades yet.")

# === DEMONSTRATION ===
print("===== STUDENT MANAGEMENT SYSTEM DEMO =====")

# 1. Add new student
print("\n1. ADDING NEW STUDENT: ")
add_student("S003", "Charlie", {"Math": 95, "Science": 97})
add_student("S004", "Sophea", {"English": 95, "Math": 90})
add_student("S001", "Alice", {"Art": 100}) 
display_all_students()

# 2. Add/update course
print("\n2. ADDING/UPDATING COURSES & GRADES")
add_update_grade('S001', 'History', 85)
add_update_grade('S003', 'Math', 92)
add_update_grade('S004', 'English', 88)
display_all_students()

# Remove a student
print("\n3. === REMOVING STUDENT ===")
remove_student('S002')
display_all_students()

# Display all students
print("\n4. === SHOW All STUDENT ===")
display_all_students()

# Display average grade
print("\n5.=== Averages ===")
show_average('S001')
show_average('S002')
show_average('S003')
show_average('S004')

# Show top students
print("\n6. === TOP STUDENT ===")
find_top_student()

print("\n===== DEMONSTRATION COMPLETE =====")
```
