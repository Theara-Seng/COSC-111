## Practice 1
You are given the following dictionary representing students and their scores on a test:
```python
scores = { 
'Alice': 85, 
'Bob': 92, 
'Charlie': 78, 
'David': 90, 
'Eve': 88 
} 
```
---
Write a Python program to: 
- Find and print the name(s) of the student(s) with the highest score. 
- Calculate and print the average score of the class. 
- Create a new dictionary that only includes students who scored at least 80. 

````python
scores = { 
'Alice': 85, 
'Bob': 92, 
'Charlie': 78, 
'David': 90, 
'Eve': 88 
} 


top_student = max(scores, key=scores.get)
print(f"Top student: {top_student} with score {scores[top_student]}")

total_score = sum(scores.values())
average_score = total_score / len(scores)
print(f"Average score: {average_score:.2f}")
    
above_80 = {student: score for student, score in scores.items() if score >= 80}
print(above_80)
````

## Practice2
You’re given: 
```
scores = { 
'Alice': 85, 'Bob': 92, 'Charlie': 78,  
'David': 90, 'Eve': 88, 'Frank': 95, 'Grace': 80 
} 
```
Write a program to: 
- Find the top 3 students with the highest scores and print their names and scores. 

````python
scores={
'Alice': 85,
'Bob': 92,
'Charlie': 78,
'David': 90,
'Eve': 88,
'Frank': 95,
'Grace': 80
}

def find_top_3_students(scores):
    sorted_scores=sorted(scores, key=scores.get)
    top_3=sorted_scores[-3:]
    return {student: scores[student] for student in top_3}
find_top_3_students(scores)
````

## Practice3

Given 
```python
data = { 
'January': {'sales': 1000, 'expenses': 400}, 
'February': {'sales': 1200, 'expenses': 500}, 
'March': {'sales': 900, 'expenses': 450} 
} 
```
Write a program to: 
- Compute and print the total sales and total expenses for the quarter. 
- Also calculate the profit (sales - expenses) for each month. 

````python
data = { 
'January': {'sales': 1000, 'expenses': 400}, 
'February': {'sales': 1200, 'expenses': 500}, 
'March': {'sales': 900, 'expenses': 450} 
} 
def total_sales(data):
    total = 0
    for month in data:
        total += data[month]['sales']
    return total
def total_expenses(data):
    total=0
    for month in data:
        total += data[month]['expenses']
    return total
def profit_each_month(data):
    profit=dict()
    for month in data:
        profit[month] = data[month]['sales'] - data[month]['expenses']
    return profit
if __name__ == "__main__":
    print("Total Sales:", total_sales(data))
    print("Total Expenses:", total_expenses(data))
    print("Profit Each Month:", profit_each_month(data))
````

## Practice4
You’re given a list of names: 
```python
names = ["Anna", "Benjamin", "Cathy", "Dann", "Eleanor", "Frank"] 
```
Write a program to create a dictionary where the keys are name lengths and the values are lists of names of that length: 
Expected Output 
```python
{ 

    4: ['Anna', 'Dann'], 
    7: ['Benjamin'], 
    5: ['Cathy', 'Frank'], 
    7: ['Eleanor']
}
```

````python
names = ["Anna", "Benjamin", "Cathy", "Dann", "Eleanor", "Frank"] 
def create_name_len(names):
    name_length = dict()
    for name in names:
        name_length[name] = len(name)
    return name_length
def reversed_dict(name_length):
    reversed_name_dict = dict()
    for name, length in name_length.items():
        if length not in reversed_name_dict:
            reversed_name_dict[length] = []
        reversed_name_dict[length].append(name)
    return reversed_name_dict
if __name__ == "__main__":
    name_length = create_name_len(names)
    reversed_name_dict = reversed_dict(name_length)
    print(reversed_name_dict)
````

## Practice5
Build a simple console-based student management system that uses Python dictionaries to store and manage data about students, their courses, and grades. 

- Store student data in a dictionary, where: 
- Each key is a student ID (or name). 
- Each value is another dictionary containing: 
    - 'name': full name 
    - 'courses': a dictionary of courses and grades. 
Example structure: 

```python
students = { 
'S001': { 
'name': 'Alice', 
'courses': {'Math': 90, 'English': 85} 
}, 
'S002': { 
'name': 'Bob', 
'courses': {'Math': 75, 'Science': 80} 
} 
} 
```
- Feature to implement 
1. Add a new student. 
2. Add or update a course & grade for a student. 
3. Remove a student by ID. 
4. Display all students with their courses and grades. 
5. Display the average grade for a given student. 
6. Find the student(s) with the highest average grade.

````python
students = { 
'S001': { 
'name': 'Alice', 
'courses': {'Math': 90, 'English': 85} 
}, 
'S002': { 
'name': 'Bob', 
'courses': {'Math': 75, 'English': 80} 
} 
} 
def get_average(scores):
    if not scores:
        return 0
    return sum(scores.values()) / len(scores)


def build_courses_dict():
    courses = {}
    while True:
        course = input("Enter course (or 'done'): ")
        if course.lower() == 'done':
            break
        try:
            grade = float(input(f"Enter grade for {course}: "))
            courses[course] = grade
        except ValueError:
            print("Invalid grade. Try again.")
    return courses

def add_student(student_id, name, courses_dict):
    if student_id not in students:
        students[student_id] = {'name': name, 'courses': courses_dict}
    else:
        print(f"Student with ID {student_id} already exists.")

def add_or_update_course(student_id, course_dict):
    if student_id in students:
        for course, grade in course_dict.items():
            students[student_id]['courses'][course] = grade  # add or update course
        print(f"Updated courses for {student_id}.")
    else:
        print(f"Student with ID {student_id} does not exist.")

def remove_student(student_id):
    target_id = student_id.lower()
    for key in list(students.keys()):
        if key.lower() == target_id:
            del students[key]
            print(f"Student ID '{key}' has been deleted.")
            return
    print(f"Student with ID '{student_id}' does not exist.")

def display_student_data():
    print("\nThis is all the student data: ")
    print("-"*30)
    for student_id, info in students.items():
        print(f"Student Id: {student_id}")
        print(f"Name: {info['name']}")
        print(f"Course and Grade: ")
        for course, grade in info["courses"].items():
            print(f"  {course}: {grade}")
        print("-"*30)

def display_student_average(name):
    for _, info in students.items():
        if name.lower() == info['name'].lower():
            student_average = get_average(info['courses'])
            print(f"{name.title()} has an average of {student_average}")
            return 
    print(f"Student {name} not found.")
    return None
def find_the_highest_grade_student():
    highest_avg = 0
    top_student = ""
    
    for _, info in students.items():
        student_average = get_average(info['courses'])
        if student_average > highest_avg:
            highest_avg = student_average
            top_student = info['name']
    
    print(f"The student with the highest average is {top_student} with a score of {highest_avg:.1f}")

def rank_students_by_average():
    ranked_list = []

    # Create a list of tuples: (name, average)
    for info in students.values():
        name = info['name']
        courses = info['courses']
        average = get_average(courses)
        ranked_list.append((name, average))

    # Sort by average descending
    ranked_list.sort(key=lambda x: x[1], reverse=True)

    # Display the results
    print("\n🏆 Student Ranking by Average Grade:")
    print("-" * 40)
    for i, (name, avg) in enumerate(ranked_list, start=1):
        print(f"{i}. {name:<15} → Average: {avg:.1f}")
    print("-" * 40)

def main():
    while True:
        print()
        print("Menu")
        print("1. Add student")
        print("2. Add or update a course and grade for students")
        print("3. Remove a student(By typing id)")
        print("4. Display all students course and grade")
        print("5. Display student average (type name)")
        print("6. Find the student with the highesh average grade")
        print("7. Display ranking (based on average).")
        print("8. Exit program\n")
        try:
            choice = int(input("Please Enter your choice (1-8): "))
            if choice not in range(1, 9):
                print("Invalid choice. Please enter a number between 1 and 7.")
                continue  # Go back to the beginning of the loop
        except ValueError:
            print("Invalid input. Please enter a valid number.")
            continue

        if choice == 1:
            student_id=input("Enter the students ID: ")
            name = input("Enter the students name: ")
            course_dict=build_courses_dict()
            add_student(student_id, name, course_dict)
        if choice ==2:
            student_id=input("Enter the students ID: ")
            course_dict=build_courses_dict()
            add_or_update_course(student_id, course_dict)
        if choice ==3:
            student_id=input("Enter the id you want to delete: ")
            remove_student(student_id)
        if choice == 4:
            display_student_data()
        if choice == 5:
            name = input("Enter the name you want to get the average: ")
            display_student_average(name)
        if choice == 6:
            find_the_highest_grade_student()
        if choice == 7:
            rank_students_by_average()
        if choice == 8:
            break


if __name__ == "__main__":
    main()
````
